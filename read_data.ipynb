```python
## Reading the data from the perspective dataset and creating the summaries

import json
import os

docs = {}
with open('dataset/theperspective/doc_new.jsonl','r') as f:
    l = f.readline()
    while(l):
        d = json.loads(l)
        docs[d['id']] = d['content']
        l = f.readline()

data = []
with open('dataset/theperspective/data.jsonl','r') as f:
    l = f.readline()
    while(l):
        d = json.loads(l)

        summary = ''

        summary+= 'Claim: ' + d['t1']+'\n'

        for i,item in enumerate(zip(d['response1'],d['favor_ids'])):
            summary+= 'Reference: ' + 'doc_' + str(item[1])+'\n'
            summary+= 'Perspective: ' + str(item[0])+'\n'
            summary+= 'Document: ' + docs[item[1]]+'\n'
        summary+='\n'

        summary+= 'Claim: ' + d['t2']+'\n'
        for i,item in enumerate(zip(d['response2'],d['against_ids'])):
            summary+= 'Reference: ' + 'doc_' + str(item[1])+'\n'
            summary+= 'Perspective: ' + str(item[0])+'\n'
            summary+= 'Document: ' + docs[item[1]]+'\n'

        data.append(summary)
        l = f.readline()
```

```python
 ## Reading the data from the perspectrumx dataset and creating the summaries    
evidence = []
with open('dataset/perspectrumx/evidence_pool.json','r') as f:
    l = f.readline()
    while(l):
        summary = ''
        e_d = json.loads(l)
        evidence.append(e_d['evidence'])
        l = f.readline()

c_data = []
with open('dataset/perspectrumx/perspectrumx.json','r') as f:
    l = f.readline()
    while(l):
        summary = ''
        c_d = json.loads(l)
        c_data.append(c_d)
        l = f.readline()

p_path = 'dataset/perspectrumx/perspective_pool_v1.0.json'
with open(p_path, 'r') as f:
    p_json_data = json.load(f)
p_data = {}
for p in p_json_data:
    p_data[p['pId']] = p['text']

data = []

for i in range(len(c_data)):
    c_d = c_data[i]
    supporting = []
    against = []
    for persp in c_d['perspectives']:
        if 'SUPPORT' in persp['stance_label']:
            eds =  persp['evidence']
            supporting.append((p_data[persp['pids'][0]],eds))
        elif 'UNDERMINE' in persp['stance_label']:
            eds = persp['evidence']
            against.append((p_data[persp['pids'][0]],eds))

    golden = 'Claim: The pespectives support the argument ' + c_d['text'] +'\n'
    if len(supporting) == 0 and len(against) == 0:
        print(i)
    for content in supporting:
        golden+= 'Perspective: ' + str(content[0])+'\n'
        golden+= 'Reference: '
        for doc_id in content[1]:
            golden+=  'doc_' + str(doc_id)+' '
        golden+= '\n'
        for doc_id in content[1]:
            golden+=  'Documents:' + evidence[doc_id]+' \n'
        golden+='\n'
    
    golden += '\nClaim: The pespectives undermine the argument ' + c_d['text'] +'\n'

    for content in against:
        golden+= 'Perspective: ' + str(content[0])+'\n'
        golden+= 'Reference: '
        for doc_id in content[1]:
            golden+=  'doc_' + str(doc_id)+' '
        golden+= '\n'
        for doc_id in content[1]:
            golden+=  'Documents:' + evidence[doc_id]+' \n'
        golden+='\n'
    
    data.append(golden)
```
